```python
import pandas as pd
```

```python
df = pd.read_csv('/content/supermarket_sales.csv')
df
```

```python
df = df.dropna()
df
```

```python
df.groupby(['City', 'Date'])['gross income'].sum()
```

```python
df.groupby(['City', 'Product line'])['Rating'].mean()
```

```python
df.groupby(['Branch', 'City', 'Gender'])['Rating'].mean()
```

```python
!pip install openpyxl

with pd.ExcelWriter('output.xlsx') as writer:
  df.to_excel(writer, sheet_name='Original')
  df.groupby(['City', 'Date'])['gross income'].sum().to_excel(writer, sheet_name='By City and Date')
  df.groupby(['City', 'Product line'])['Rating'].mean().to_excel(writer, sheet_name='City Product Line Rating')
  df.groupby(['Branch', 'City', 'Gender'])['Rating'].mean().to_excel(writer, sheet_name='Branch City Gender Rating')
```
